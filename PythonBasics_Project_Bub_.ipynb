## Задание на проект

**Проект "Кредит"** (Ольга Бубнова)

Необходимо рассчитать график платежей по кредиту на сумму 500000 руб. на срок 36 месяцев под 12% годовых. Создайте функцию или метод класса, которые будут на вход принимать значения суммы кредита, срок кредита и процентную ставку, а возвращать список со значениями аннуитетного платежа и остатка долга после платежа. Для расчета суммы аннуитетного платежа используйте формулу:

\begin{align}
R = \frac{S \frac{P}{12 \cdot 100}}{1-(1+\frac{P}{12 \cdot 100})^{-T}} \\
\end{align}

где $R$ - сумма платежа, руб.; $S$ - сумма кредита, руб.; $P$ - процентная ставка, %; $T$ - количество платежей, шт.

Оцените также переплату за кредит.

```
class Credit:
    def __init__(self, cr_amount, cr_term, int_rate):
        self.S=cr_amount
        self.P=int_rate
        self.T=cr_term
        self.l=[0]
        self.b=[0]
    def calc (self):
        self.k=self.P/(12*100)
        self.payment=(self.S*self.k)/(1-(1+self.k)**(-self.T))
        self.amount=self.payment*self.T
        self.overpay=self.amount-self.S
        for i in range(1, self.T+1, 1):
            self.l.append(self.payment)
            self.b.append(self.amount-sum(self.l))
            print('Ежемесячный платеж №'+str(i)+': %.2f' %self.l[i], 'руб. Остаток долга после платежа: %.2f руб.' %self.b[i])
        print ('Переплата по кредиту: %.2f руб.' %self.overpay)
result=Credit(500000,36,12)
result.calc()
```
